```python
import pandas as pd
import numpy as np
import datetime as dt
import requests
# import gmaps
import os
# Import API key
from api_keys import g_key
```

```python
data_set = 'data/Crew Data.csv'
input_data = pd.read_csv(data_set) 

input_data
```

```python
Hometown = []
for row in input_data.index:
    name = input_data.at[row, 'Name']
    home = input_data.at[row, 'Hometown']
    Hometown.append(home)
Hometown_set = set(Hometown)
print(Hometown)
print(Hometown_set)
```

```python
finalList = []
cleanedList = [x for x in Hometown_set if str(x) != 'nan']
print(cleanedList)
for s in cleanedList:
    if "," in s:
        list = [s,s]
        finalList.append(list)
    else:
        new = s + ', UK'
        list = [s,new]
        finalList.append(list)
print(finalList)
```

```python
output = []
url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json?"

for home in finalList:
    print(f'Retrieving Results for Home | {home[1]}') 
    
    parameter = (f"input={home[1]}&inputtype=textquery&fields=formatted_address,name,geometry&key={g_key}")

    
    try:

#Get the response from the API request and capture the lat/lng data values from the request
        response = requests.get(url + parameter).json()  
        print(response)

#extract results
        home_Lat = response['candidates'][0]['geometry']['location']['lat']
        home_Lng = response['candidates'][0]['geometry']['location']['lng']
        
        results = [home[0], home[1], home_Lat, home_Lng]
        print (results)
        output.append(results)
#Generate an exception message if passenger data not found
    except Exception:
        print(f'{home} city not found. Skipping...')
print(output)
```

```python
for hometown in output:
    town = hometown[0]
    lat = hometown[2]
    lng = hometown[3]
    
    for row in input_data.index:
        home = input_data.at[row, 'Hometown']
    
        if town == home:
            
           input_data.at[row, 'Home_Lat'] = lat
           input_data.at[row, 'Home_Lng'] = lng
        
input_data 
```

```python
input_data.to_csv('data/CrewData_LatLong.csv')        
print(input_data) 
```
